# Preprocess

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
import re
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
# t = "to be or not to be"
# from nltk.util import ngrams
# tokens = nltk.word_tokenize(t)
# bigrm = nltk.bigrams(tokens)
# v= [' '.join(e) for e in ngrams(tokens, 2)]
# v
```

```python
#from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from string import punctuation 
from nltk.corpus import stopwords
#nltk.download('stopwords')
from nltk.stem.porter import *
from nltk.stem import WordNetLemmatizer

preprocessed=[]
for txt in tweets:
  txt = txt.lower() # convert text to lower-case
  txt = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', '', txt) # remove URLs
  txt = re.sub('@[^\s]+', '', txt) # remove usernames
  txt = re.sub(" \d+", " ", txt) #remove digits
  txt = re.sub(r'#([^\s]+)', r'\1', txt) # remove the # in #hashtag
  txt = re.sub(r'[^\w\s]', '', txt) #remove punctuation
  txt = deEmojify(str(txt)) # remove emojy
  txt = word_tokenize(txt)
  # v = [' '.join(e) for e in ngrams(txt, 2)]
  # final = [word for word in v if not word in stopwords.words()] #stop words 
  preprocessed.append(txt)
```

```python
data = pd.DataFrame(preprocessed).fillna('0')
data
```

# Translation and Lemmatization an Removing stop words

```python
# !pip install googletrans==3.1.0a0
```

```python
import googletrans
from googletrans import Translator
```

```python
translator = Translator()
lemmatizer = WordNetLemmatizer()
```

```python
# (translator.translate('Mit√§')).text
```

```python
# bidata=[]
# for sentence in data.values.tolist():
#   v = [' '.join(e) for e in ngrams(sentence, 2)]
#   bidata.append(v)
```

```python
pdata=[]
for sentence in data.values.tolist():
  tmp = []
  for word in sentence:
    translated = translator.translate(word)
    final = lemmatizer.lemmatize(translated.text)
    if len(final)>=3:
      tmp.append(final)
  pdata.append(tmp)
```

# Topic Modeling

```python
from bertopic import BERTopic
```

```python
Data=[]
for vec in pdata:
  text=''
  for w in vec:
    if w != '0' and w :
      text = text + w + ' '
  Data.append(text)
```

```python
Data[:5]
```

```python
tokenized_data = []
for text in Data:
  tokens = word_tokenize(text)
  tmp=[]
  for token in tokens:
    if token not in stopwords.words() and len(token)>=3:
      tmp.append(token)
  tokenized_data.append(tmp)
```

```python
clean=[]
pattern = r'[0-9]'
for vec in tokenized_data:
  text=''
  for w in vec:
    if w != '0' and w :
      text = text + w + ' '
      text = re.sub(pattern, "", text)
  clean.append(text)
```

```python
pd.DataFrame(clean).head()
```

```python
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)
topics, probs = topic_model.fit_transform(clean)
```

```python
topic_model.update_topics(clean, topics, n_gram_range=(1, 2))
```
